```sql
%sql
select Pais, sum(Preco) as total_vendido from vinhos_mundo
where Preco > 0
group by pais
order by total_vendido desc
limit 10
```

```sql
%sql
select pais, variedade, sum(Preco) as total_vendido from vinhos_mundo
where Preco > 0
group by pais, variedade
order by total_vendido desc
limit 10
```
